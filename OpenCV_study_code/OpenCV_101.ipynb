```python
import numpy as np
img = np.zeros((512, 512, 3),np.uint8)
plt.imshow(img)
```

```python
img = cv2.line(img, (0,0),(511,511),(255,0,0),5)
plt.imshow(img)
plt.show()
```

```python
imh = cv2.rectangle(img,(400,0),(510,128),(0,255,0),5)
plt.imshow(img)
plt.show()
```

```python
img = cv2.circle(img, (450,50), 50, (0,0,255),-1)
plt.imshow(img)
plt.show()
```

```python
img = cv2.circle(img, (50,450), 50, (255,255,255),5)
plt.imshow(img)
plt.show()
```

```python
img = cv2.ellipse(img,(256,256),(150,30),0,0,180,(0,255,0),-1)
plt.imshow(img)
plt.show()
```

```python
pts = np.array([[10,5],[20,30],[70,20],[50,10]],np.int32)
print(pts.shape)
pts = pts.reshape((-1,2,1))
print(pts.shape)
img = cv2.polylines(img,[pts],True, (172,200,255),4)
plt.imshow(img)
plt.show()
```
